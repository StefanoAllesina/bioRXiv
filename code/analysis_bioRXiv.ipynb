```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
data = pd.read_csv('bioRXiv.csv')
```

```python
data
```

```python
# transform to datetime
data.date = pd.to_datetime(data.date)
```

```python
# plot submission by day
submissions_by_day = data.groupby(['date']).size()
```

```python
submissions_by_day.plot()
```

```python
# plot cumulative submissions
submissions_by_day.cumsum().plot()
```

```python
# use log for y axis
submissions_by_day.cumsum().plot(logy = True)
```

```python
# average number of authors by category
z = pd.DataFrame(data[['category','num_authors']]).boxplot(by = 'category')
plt.xticks(rotation = 90)
plt.yscale('log')
```
